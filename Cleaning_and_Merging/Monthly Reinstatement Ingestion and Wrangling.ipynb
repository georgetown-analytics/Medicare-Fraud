```python
#Reading dataframe back in
reinstatements = pd.read_csv('Data/Reinstatements/Full_Reinstatements.csv', dtype=object)
```

```python
#Confirming dataframe read in correctly
reinstatements.head()
```

```python
#Creating a column with the year only, instead of exact date, to easily filter the data by year
reinstatements['EXCLYear'] = reinstatements['EXCLDATE'].str[0:4]
```

```python
#Creating a column for the reinstatement year as well
reinstatements['REINYear'] = reinstatements['REINDATE'].str[0:4]
```

```python
#Checking columns
reinstatements.columns
```

```python
#Dropping unnneeded columns
reinstate = reinstatements.drop(['GENERAL', 'UPIN', 'EXCLDATE', 'REINDATE', 'WAIVERDATE', 'WVRSTATE'], axis = 1)
```

```python
reinstate.head()
```

```python
#Checking to ocnfirm few null values, most likely in the name fields
reinstate.isnull().sum()
```

```python
#Looking at number of entries that have a valid NPI
reinstate['NPI'].value_counts()
```

```python
#Full LEIE dataset needed cleaning for the exclusion type so looking at the values here
reinstate['EXCLTYPE'].value_counts()
```

```python
#Cleaning up EXCLTYPE as there are spaces at the end of some, making them look like a separate code but unsure how many spaces are at the end so harmonizing it
reinstate['excl_type'] = reinstate['EXCLTYPE'].str[0:7]
```

```python
#Merging the one duplicate code (1128a1) to be the same
reinstate.replace({'excl_type': {'1128a1 ': '1128a1'}}, inplace = True)
```

```python
#confirming this merged the two duplicates
reinstate['excl_type'].value_counts()
```

```python
#Dropping the unclean EXCLTYPE column
reinstate.drop(columns = ['EXCLTYPE'])
```

```python
#Looking at the range of exclusion dates
reinstate['EXCLYear'].value_counts()
```

```python
#Converting columns to numeric values
reinstate['EXCLYear'] = pd.to_numeric(reinstate['EXCLYear'])
```

```python
reinstate['NPI'] = pd.to_numeric(reinstate['NPI'])
```

```python
#Filtering to only exlusions that occurred after 2011
ge_2011 = reinstate['EXCLYear'] > 2010
reinstate_2011 = reinstate[ge_2011]
```

```python
#Looking at the number of exclusions years included in data 
fig = plt.figure(figsize=(7,10))
ax = fig.add_subplot(211)
ax.hist(reinstate_2011['EXCLYear'], range=(reinstate_2011['EXCLYear'].min(), reinstate_2011['EXCLYear'].max()))
plt.title('Number of Medicare Exclusions Per Year')
plt.xlabel('Year of Exclusion')
plt.ylabel('Number of Exclusions')
plt.show()
```

```python
#Saving filtered data as new csv
reinstate_2011.to_csv('Data/reinstatements_2011.csv', index=False)
```

```python
#Reading saved data back in to create two datasets, one with NPI and one without
df = pd.read_csv('Data/reinstatements_2011.csv')
```

```python
#Creating a dataset without NPIs
no_NPI = df['NPI'] == 0
df_no_NPI = df[no_NPI]
```

```python
#Confirming all have a 0 value for NPI
df_no_NPI['NPI'].value_counts()
```

```python
df_no_NPI.shape
```

```python
#Saving the no NPI file as well to wrangle further if needed
df_no_NPI.to_csv('Data/reinstatements_noNPI.csv', index=False)
```

```python
#Creating a dataset with NPIs
NPI = df['NPI'] > 0
df_NPI = df[NPI]
```

```python
#confirming no 0 values for NPI
df_NPI['NPI'].value_counts()
```

```python
#Sorting to most recent exclusion
df_NPI.sort_values(by = 'EXCLYear', ascending=False)
```

```python
#Remove the duplicate row
df_NPI_dedupped = df_NPI.drop_duplicates(subset = 'NPI', keep = 'first')
```

```python
#Confirming one duplicate dropped
df_NPI_dedupped['NPI'].value_counts()
```

```python
#Saving as new csv to merge with Provider data
df_NPI_dedupped.to_csv('Data/reinstatements_NPI.csv', index=False)
```
